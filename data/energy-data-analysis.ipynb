## Look at the emission

```python
simple = df[['Primary Property Type - Portfolio Manager-Calculated','Total GHG Emissions (Metric Tons CO2e)']]
```

```python
simple = simple.replace('Not Available', np.nan)
simple['Total GHG Emissions (Metric Tons CO2e)'] = simple['Total GHG Emissions (Metric Tons CO2e)'].astype(float)
```

```python
p = simple.groupby('Primary Property Type - Portfolio Manager-Calculated')['Total GHG Emissions (Metric Tons CO2e)'].sum().reset_index()
```

```python
p['Percentage'] = 100 * p['Total GHG Emissions (Metric Tons CO2e)']  / p['Total GHG Emissions (Metric Tons CO2e)'].sum()
```

```python
p.sort_values(by=['Percentage'],ascending=False)
```

```python
p = p.rename(columns={"Primary Property Type - Portfolio Manager-Calculated": "type", \
                  "Total GHG Emissions (Metric Tons CO2e)": "ghg_emission", \
                  "Percentage":"percentage"})
```

```python
type_number = simple['Primary Property Type - Portfolio Manager-Calculated'].value_counts().to_frame().reset_index()
```

```python
type_number  = type_number.rename(columns={"index": "type", \
                  "Primary Property Type - Portfolio Manager-Calculated": "number"})
```

```python
type_number.head()
```

```python
type_viz = pd.merge(p, type_number, on=['type'])
type_viz
```

```python
type_viz['by_number'] = type_viz['ghg_emission'] / type_viz['number']
type_viz.sort_values(by=['ghg_emission'],ascending = False).head(20)
```

```python
type_viz.to_csv("type_vis.csv",index=False)
```

## Look at the housing emission particularly

```python
df_h.head()
```

```python
h_emission = df_h[['Borough','NYC Borough, Block and Lot (BBL)','ENERGY STAR Score','Site EUI (kBtu/ft²)','Total GHG Emissions (Metric Tons CO2e)','Total GHG Emissions Intensity (kgCO2e/ft²)']]
```

```python
h_emission = h_emission.rename(columns={"Borough": "borough", \
                  "NYC Borough, Block and Lot (BBL)": "BBL", \
                  "ENERGY STAR Score": "energy_star_core", \
                  "Site EUI (kBtu/ft²)":"EUI",
                  'Total GHG Emissions (Metric Tons CO2e)':"emission_ton",
                  'Total GHG Emissions Intensity (kgCO2e/ft²)':"emission_intensity_kg/ft2"})
```

```python
h_emission.borough.value_counts()
```

```python
h_emission = h_emission.replace('Not Available', np.nan)
h_emission[['energy_star_core','EUI','emission_ton','emission_intensity_kg/ft2']] = h_emission[['energy_star_core','EUI','emission_ton','emission_intensity_kg/ft2']].astype(float)
```

```python
h_emission['square'] = (h_emission.emission_ton * 1000 / h_emission['emission_intensity_kg/ft2']).round(1)
h_emission['standard'] = 0.00675
h_emission['penalty'] = (h_emission.emission_ton - h_emission['standard'] * h_emission['square']) * 268
h_emission['penalty'] = h_emission['penalty'].round(1)
h_emission.sample(10)
```

```python
penalty_number = h_emission[h_emission.penalty > 0]
```
